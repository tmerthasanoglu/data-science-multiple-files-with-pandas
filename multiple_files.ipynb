## Merge AlÄ±ÅŸtÄ±rmasÄ±

```python
a_df = pd.DataFrame({
    'Country': ['Germany', 'France', 'Belgium', 'Finland'],
    'Population (M)': [82.8, 67.2, 11.4, 5.5],
    'Capital': ['Berlin', 'Paris', 'Brussels', 'Helsinki']
})
a_df
```

```python
b_df = pd.DataFrame({
    'Country': ['Germany', 'France', 'Belgium', 'Canada'],
    'HDI': [0.936, 0.901, 0.916, 0.926]
})
b_df
```

### Ä°Ã§ BirleÅŸtirme AlÄ±ÅŸtÄ±rmasÄ±

```python
inner_merged_df=pd.DataFrame.merge(a_df,b_df,on="Country")
inner_merged_df
```

#### Kodunu kontrol et

```python
from nbresult import ChallengeResult

result = ChallengeResult('inner_merge',
    inner_merged_shape=inner_merged_df.shape,
    inner_merged_nulls=sum(inner_merged_df.isnull().sum())
)
result.write()

print(result.check())
```

### Sol BirleÅŸtirme AlÄ±ÅŸtÄ±rmasÄ±

```python
left_merged_df=pd.DataFrame.merge(a_df,b_df,on="Country",how="left")
left_merged_df
```

#### Kodunu kontrol et


```python
from nbresult import ChallengeResult

result = ChallengeResult('left_merge',
    left_merged_shape=left_merged_df.shape,
    left_merged_nulls=sum(left_merged_df.isnull().sum())
)
result.write()

print(result.check())
```

### SaÄŸ BirleÅŸtirme AlÄ±ÅŸtÄ±rmasÄ±


```python
right_merged_df=pd.DataFrame.merge(a_df,b_df,on="Country",how="right")
right_merged_df
```

#### Kodunu kontrol et

```python
from nbresult import ChallengeResult

result = ChallengeResult('right_merge',
    right_merged_shape=right_merged_df.shape,
    right_merged_nulls=sum(right_merged_df.isnull().sum())
)
result.write()

print(result.check())
```

### DÄ±ÅŸ BirleÅŸtirme AlÄ±ÅŸtÄ±rmasÄ±

## Olympic Sports and Medals, 1896-2014

```python
countries_df=pd.read_csv("data/dictionary.csv")
summer_df=pd.read_csv("data/summer.csv")
winter_df=pd.read_csv("data/winter.csv")
```

```python
countries_df
```

```python
summer_df
```

### Verileri BirleÅŸtirme


```python
# 1) Summer merge (Country kodu â†” Code)
summer_countries_df = pd.merge(
    summer_df,
    countries_df,
    left_on="Country",
    right_on="Code",
    how="inner"
)

# Country_x: kod, Country_y: Ã¼lke adÄ± -> isimleri test formatÄ±na Ã§evir
summer_countries_df = summer_countries_df.rename(
    columns={"Country_x": "Code", "Country_y": "Country"}
)

# Season ekle
summer_countries_df["Season"] = "Summer"
```

```python
# 2) Winter merge (aynÄ±)
winter_countries_df = pd.merge(
    winter_df,
    countries_df,
    left_on="Country",
    right_on="Code",
    how="inner"
)

winter_countries_df = winter_countries_df.rename(
    columns={"Country_x": "Code", "Country_y": "Country"}
)

winter_countries_df["Season"] = "Winter"
```

```python
all_df = pd.concat([summer_countries_df, winter_countries_df], ignore_index=True)
expected_cols = [
    "Year","City","Sport","Discipline","Athlete",
    "Country","Gender","Event","Medal",
    "Code","Population","GDP per Capita","Season"
]
all_df = all_df.loc[:, ~all_df.columns.duplicated()]
all_df = all_df[expected_cols]
```

#### Kodunu kontrol et


```python
from nbresult import ChallengeResult

result = ChallengeResult('all_df',
    all_df_shape=all_df.shape,
    all_df_columns=set(all_df.columns)
)
result.write()

print(result.check())
```

### En BaÅŸarÄ±lÄ± Ãœlkeler Analizi

```python
top_10_df = (
    all_df.loc[all_df["Year"] >= 1984]                 # boolean indexing
          .groupby("Country")["Medal"]                 # gruplama
          .count()                                     # madalya say
          .sort_values(ascending=False)                # sÄ±ralama
          .head(10)                                    # ilk 10
          .to_frame(name="Medal Count")                # tek sÃ¼tunlu DF
)
top_10_df
```
