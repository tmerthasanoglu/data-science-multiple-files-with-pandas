```python
import numpy as np
import pandas as pd
import matplotlib
```

## Merge AlÄ±ÅŸtÄ±rmasÄ±

```python
a_df = pd.DataFrame({
    'Country': ['Germany', 'France', 'Belgium', 'Finland'],
    'Population (M)': [82.8, 67.2, 11.4, 5.5],
    'Capital': ['Berlin', 'Paris', 'Brussels', 'Helsinki']
})
a_df
```

```python
b_df = pd.DataFrame({
    'Country': ['Germany', 'France', 'Belgium', 'Canada'],
    'HDI': [0.936, 0.901, 0.916, 0.926]
})
b_df
```

### Ä°Ã§ BirleÅŸtirme AlÄ±ÅŸtÄ±rmasÄ±

```python
inner_merged_df=pd.DataFrame.merge(a_df,b_df,on="Country")
inner_merged_df
```

#### Kodunu kontrol et

```python
from nbresult import ChallengeResult

result = ChallengeResult('inner_merge',
    inner_merged_shape=inner_merged_df.shape,
    inner_merged_nulls=sum(inner_merged_df.isnull().sum())
)
result.write()

print(result.check())
```

### Sol BirleÅŸtirme AlÄ±ÅŸtÄ±rmasÄ±

```python
left_merged_df=pd.DataFrame.merge(a_df,b_df,on="Country",how="left")
left_merged_df
```

#### Kodunu kontrol et


```python
from nbresult import ChallengeResult

result = ChallengeResult('left_merge',
    left_merged_shape=left_merged_df.shape,
    left_merged_nulls=sum(left_merged_df.isnull().sum())
)
result.write()

print(result.check())
```

### SaÄŸ BirleÅŸtirme AlÄ±ÅŸtÄ±rmasÄ±


```python
right_merged_df=pd.DataFrame.merge(a_df,b_df,on="Country",how="right")
right_merged_df
```

#### Kodunu kontrol et

```python
from nbresult import ChallengeResult

result = ChallengeResult('right_merge',
    right_merged_shape=right_merged_df.shape,
    right_merged_nulls=sum(right_merged_df.isnull().sum())
)
result.write()

print(result.check())
```
